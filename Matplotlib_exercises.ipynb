```python
import matplotlib.pyplot as plt
import math
```

```python
x = range(-50, 50)

# for loop. Loop through each element and perform some math

y = []
for num in x:
    y.append((num**2 - num + 2))
```

```python
x = range(-10, 10)
y = [(x**2 -x + 2) for x in x]


plt.plot(x,y)
```

```python
# generate data
x = range(-100, 100)
y = [(x**2 - x + 2) for x in x]


# Plot x and y
plt.plot(x,y)

# add title
plt.title('$ y = x^2 - x +2$')

# add labels
plt.xlabel('$x$')
plt.ylabel('$y$')

# add annotation
plt.annotate('Origin', xy = (0,0), xytext = (0, 3000), arrowprops ={})
%config InlineBackend.figure_format = 'png'

# save the figure
plt.savefig('figure1.png')
```

```python
# generate data
x1 = range(0,100)
y1 = [math.sqrt(n) for n in x1]

# plot the data
plt.plot(x1,y1)

# customize the chart
plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)

plt.grid(True, ls = '--');
```

```python
# generate data
x2 = range(0,20)
y2 = [n**3 for n in x2]

#plot data
plt.plot(x2,y2)

#customize the chart
plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--');
```

```python
# generate data
x3 = [x * 0.1 for x in range(0,150)]
y3 = [2 ** n for n in x3]

# plot data
plt.plot(x3,y3)

#customize the chart
plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.grid(True, ls = '--')
# plt.xticks(range(0, 16))
;
```

```python
# specify range of x values
x4 = [x * 0.1 for x in range(0,150)]
y4 = [1/(n+1) for n in x4]


# plot the data
plt.plot(x4,y4)


# Customize the plot
# Add title
plt.title('$ y = 1/(x+1)$')

# Add x and y labels
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')


plt.grid(True, ls = '--')

plt.show()
```

```python
plt.figure(figsize = (10,6))

#subplot 1
plt.subplot(221)

plt.plot(x1,y1)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)
plt.grid(True, ls = '--')

#subplot 2
plt.subplot(222)

plt.plot(x2,y2)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--')


#subplot 3
plt.subplot(223)

plt.plot(x3,y3)
plt.grid(True, ls = '--')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.title('$y = 2^x$')

#subplot 4
plt.subplot(224)

plt.plot(x4,y4)
plt.title('$ y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$1/x$')
plt.grid(True, ls = '--')
plt.xlim(0,15)
plt.ylim(0)
plt.annotate("some text", xy = (2,0.1), xytext = (2, 0.6), arrowprops = {});


plt.tight_layout()
```

```python
plt.figure(figsize = (8,6))

# plot the data
plt.plot(x1, y1, c='#30a2da', label='$\sqrt{x}$', alpha=1.0)
plt.plot(x2, y2, c='orange', label='$x^{3}$', alpha=.8)
plt.plot(x3, y3, c= (1,0,1), label='$2^{x}$', alpha=.8)
plt.plot(x4, y4, c= '#444444', label='$1/(x+1)$', alpha=.8)

# adjust xlim and ylim
plt.ylim(0, 10)
plt.xlim(0, 10)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('My Functions')
plt.legend()

plt.show()
```
