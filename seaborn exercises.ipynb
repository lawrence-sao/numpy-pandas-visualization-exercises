```python
swiss = data('swiss')
```

```python
swiss['is_catholic']=swiss.Catholic>=70
```

```python
swiss
```

```python
sns.relplot(x='Fertility', y='is_catholic', data=swiss)

##-- ANSWER Provinces that are Catholic seem to have higher fertility on average than non catholic provinces, their numbers start around 60% and plot between 60-90% compared to other provinces which are more spread out and start much lower (sub 40%)
```

```python
correlation_swiss = swiss.corr()
correlation_swiss = correlation_swiss.drop(columns=["Fertility","is_catholic"])
correlation_swiss = correlation_swiss.head(1)
correlation_swiss
sns.heatmap(correlation_swiss)

##Answer: Catholicism correlates most strongly with Fertility rates 
```
