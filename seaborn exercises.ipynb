```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
iris = sns.load_dataset('iris')

iris.dtypes
```

```python
#What does the distribution of petal lengths look like?

sns.histplot(iris.petal_length)
plt.show()
```

```python
#Is there a correlation between petal length and petal width? Use http://guessthecorrelation.com/ as a hint to how we could visually determine if a correlation exists between two numeric columns. 
    
sns.relplot(x='petal_width', y='petal_length', data=iris)

#iris.head() there is positive correlation
```

```python
sns.jointplot(data=iris, x='sepal_width', y='species')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='species')
```

```python

Would it be reasonable to predict species based on sepal width and sepal length? 
# yes it would be reasonable for setosa but not the other species  
Which features would be best used to predict species?
# Petal length and width is better for predicting all thre species
```

```python
sns.relplot(x='petal_length', y='petal_width', hue='species', data=iris)
```
